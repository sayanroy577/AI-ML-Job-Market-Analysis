```python
#Importing libraries.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importing data.
df=pd.read_csv('ai_job.csv')
df.head(5)
```

```python
#Checking information of datasets.
df.info()
```

```python
#Show all columns of dataset.
df.columns
```

```python
#Provide statistical information of datasets.
df.describe()
```

```python
#1.What is the distribution of salaries?
sns.histplot(x='salary_usd',data=df,bins=20,kde=True)
plt.title('Distribution of Salaries')
```

#The distribution is right skewed. Which means outliers are present as high paying jobs.

```python
#2.How does experience level impact salary?
sns.boxplot(x='experience_level',y='salary_usd',data=df)
plt.title('Experience Level Distribution')
```

#Executives are drawing more salary than rest employees.

```python
#3.Visualize the salary distribution by job title
plt.figure(figsize=(10,3))
sns.violinplot(x='job_title',y='salary_usd',data=df)
plt.xticks(rotation=90)
plt.title('Salary distribution by job title')
```

#Senior AI roles like Principal Data Scientist and Head of AI command the highest salaries (up to $400k), reflecting significant experience premiums in the field.

```python
#4.Analyze the relationship between salary & years of experience.
sns.scatterplot(x='years_experience', y='salary_usd', data=df)
plt.title('Years of Experience vs Salary')
```

#As experiences increases, salary also increases.

```python
#5.Visualize the top 5 companies with the highest number of job postings.
top_companies = df['company_name'].value_counts().head(5)
plt.figure(figsize=(10,4))
sns.barplot(x=top_companies.index, y=top_companies.values,palette='viridis')
plt.xticks(rotation=90)
plt.title('Top 5 companies with highest job postings')
```

#These 5 companies are hiring more.

```python
#6.Does remote work influences Salary?
ax=sns.barplot(x='remote_ratio',y='salary_usd',data=df)
ax.bar_label(ax.containers[0])
plt.title('Remote Ratio vs Salary')
```

#Fully remote role sometimes offer higher salary

```python
#7.Is benefit score influencing salary?
plt.figure(figsize=(10,4))
sns.scatterplot(x='benefits_score', y='salary_usd', data=df)
plt.title('Benefits Score vs Salary')
```

#Mild positive correlation i.e. better benefits may accompany higher pay.

```python
#8.What is the year of experience of majority employees?
sns.boxplot(x='years_experience',data=df)
plt.title('Years of Experience')
```

#Majority of employees have experience between 2-10 years.
